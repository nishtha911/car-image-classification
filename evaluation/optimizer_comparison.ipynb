```python
import json
import os
import matplotlib.pyplot as plt
```

```python
LOG_DIR = "../logs"
```

```python
def load_history(name):
    
    path = os.path.join(LOG_DIR, f"{name}_history.json")
    
    with open(path, "r") as f:
        history = json.load(f)
    
    return history
```

```python
adam = load_history("adam")

sgd = load_history("sgd")
rmsprop = load_history("rmsprop")
adagrad = load_history("adagrad")
```

```python
plt.figure(figsize=(10,6))

plt.plot(adam["val_loss"], label="Adam")

# uncomment later
plt.plot(sgd["val_loss"], label="SGD")
plt.plot(rmsprop["val_loss"], label="RMSprop")
plt.plot(adagrad["val_loss"], label="Adagrad")

plt.title("Optimizer Comparison — Validation Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()

plt.savefig("../logs/optimizer_comparison_loss.png")

plt.show()
```

```python
plt.figure(figsize=(10,6))

plt.plot(adam["val_mae"], label="Adam")

# uncomment later
plt.plot(sgd["val_mae"], label="SGD")
plt.plot(rmsprop["val_mae"], label="RMSprop")
plt.plot(adagrad["val_mae"], label="Adagrad")

plt.title("Optimizer Comparison — Validation MAE")
plt.xlabel("Epoch")
plt.ylabel("MAE")
plt.legend()

plt.savefig("../logs/optimizer_comparison_mae.png")

plt.show()
```
